```python
import numpy as np
import pandas as pd
import statsmodels.api as sm
from statsmodels.tsa.arima_process import arma_generate_sample
import matplotlib
import seaborn as sns
%matplotlib inline
matplotlib.rcParams['figure.figsize'] = (15, 6)

np.random.seed(1)

x1 = arma_generate_sample(ar=[0.999], ma=[0.9], nsample=100) + 100
y = 1.2 * x1 + np.random.randn(100)

y[71:100] = y[71:100] + 10
data = pd.DataFrame(np.array([y, x1]).T, columns=["y","x1"])
```

Visualize data

```python
data.plot()
```

Plot data before intervention

```python
prePeriod_df= data[:69]
prePeriod_df.plot()
```

```python
X = prePeriod_df.x1
```

```python
X = sm.add_constant(X)
```

```python
Y = prePeriod_df.y
```

```python
results = sm.OLS(Y, X).fit()
```

```python
print(results.summary())
```

```python
postPeriod_df= data[71:99]
```

```python
predictions = results.params[0] +postPeriod_df.x1*results.params[1]
```

```python
data["y*"] = predictions
```

```python
data.plot()
```
